```
#17
k1=input("Enter key 1")
k2=input("Enter key 2")
k3=input("Enter key 3")
k4=input("Enter key 4")
v1=input("Enter the value of key 1")
v2=input("Enter the value of key 2")
v3=input("Enter the value of key 3")
v4=input("Enter the value of key 4")
d={}
d[k1]=v1
d[k2]=v2
d[k3]=v3
d[k4]=v4
l=list(d.items())
l.sort()
print("ascending order:",l)
l=list(d.items())
l.sort(reverse= True)
print("descending order",l)
```

```
#17
x=input("enter the first string : ")
y=input("enter the second string : ")
def swapch(x,y): 
  a = y[:1] + x[1:]
  b = x[:1] + y[1:]
  return a + ' ' + b
print(swapch(x,y))
```

```
x=[]
a=int(input("enter limit for list1 : "))
for i in range(1,a+1):
  a=int(input())
  x.append(a)
print(x)
y=[]
b=int(input("enter limit for list2 : "))
for i in range(1,b+1):
  b=int(input())
  y.append(b)
print(y)
if (len(x)==len(y)):
  print("list1 and list2 are of same length i.e.",len(x))
else:
  print("list1 and list2 are of different length.")
if (sum(x)==sum(y)):
  print("list1 and list2 sums to the same value i.e.",sum(x))
else:
  print("list1 and list2 sums up to different values.")
def common(x, y): 
    p=set(x) 
    q=set(y) 
    if len(p.intersection(q)) > 0: 
        return(p.intersection(q))   
    else: 
        return("no common elements")  
print("value that occur in both lists is/are :",common(x,y)) 
```
